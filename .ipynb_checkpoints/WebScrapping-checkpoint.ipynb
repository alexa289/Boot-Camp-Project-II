```python
def scrape_wine():
    browser = webdriver.Chrome('chromedriver.exe')
    news = {}
    url = "https://www.winemag.com/?s=&drink_type=wine"
    browser.get(url)
    time.sleep(1)
    
    html = browser.page_source 
    soup = BeautifulSoup(html, "html.parser")

    for x in range(1, 6):

        html = browser.page_source
        soup = BeautifulSoup(html, 'html.parser')

        
        reviews = soup.find_all('li', class_='review-item')
        for review in reviews:
            #title = review.find("div", class_= 'title')
#             print(title)
#             print('page:', x, '-------------')
#             print(review)

    # ##########################################################        
    # Loop through returned results
    #     for result in results:
                # Error handling
            try:
                # Identify and return title of listing
                title = review.find('div', class_='title').text
                rating = review.find('span', class_='rating').text
                appellation = review.find('span', class_='appellation').text 
                price = review.find('span', class_='price').text 
                badge = review.find('span', class_='badge').text
                
                # Run only if title, price, and link are available
                if (title and rating):
                    # Print results
                    print('-------------')
                    print(title)
                    print(rating)
                    print(appellation)
                    print(price)


                    # Dictionary to be inserted as a MongoDB document
                    post = {
                        'title': title,
                        'rating': rating,
                        'appellation': appellation,
                        'price': price,
                        'badge': badge
                        }

                    collection.insert_one(post)

            except Exception as e:
                print(e)        

        #     browser.click_link_by_partial_text('Next')


#     return wine 
scrape_wine()


#     news["news_title"] = soup.find("div", class_="content_title").get_text()
#     news["news_p"] = soup.find("div", class_="article_teaser_body").get_text()

#     return news
# scrape_news()
```
